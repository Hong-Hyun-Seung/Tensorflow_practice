gradient descent algorithm은 cost함수를 미분해서 global minimum으로 따라 내려가는것인데, 이를 위해 "최종 값을 cost함수에 넣은것"(ex) -최종값 log 최종값)을 미분해야 한다. 

그렇지만 신경망에서는 미분값을 도출하기 매우 어렵다

최적해를 찾기 위해서는 Y의 "미분"이 필요한데 (bc/!최종 Y의 값을 넣은 cost함수를 "미분"한 값! 앞에 알파를 조절하는 것이 optimizer의 역할), 이때 Y의 미분에 각각의 input이 얼마나 영향을 미치는지 (ex)dY/dX1) 알아야 하고(->얼마나 영향을 미치는지에 따라서 w조정), 이때 각각의 미분값을 구하기 위해서는 너무 많은 계산량을 요구한다
